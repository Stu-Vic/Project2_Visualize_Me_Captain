```python
#Import Dependencies
import pandas as pd
```

```python
#Create reference to CSV file
csv_path = "ArianaGrandeTweets.csv"

#Import the CSV into a pandas DataFrame
AGrande_Tweets_df = pd.read_csv(csv_path, low_memory = False)
AGrande_Tweets_df
```

```python
#Show the 'shape' of the dataframe 
shape = AGrande_Tweets_df.shape
print(f"The dataframes 'shape' is represented as: {shape}")
```

##### COLLECT ONLY REQUIRED COLUMNS

```python
#Select only columns we need
AGrande_Tweets_df = AGrande_Tweets_df[["Username", "User Id", "Tweet Type", "Tweet Content", "Client", "Retweets received",
                                     "Likes received"]]

#Show the 'shape' of the dataframe
shape = AGrande_Tweets_df.shape
print(f"The dataframes 'shape' is represented as: {shape}")
```

```python
#Print the dataframe
AGrande_Tweets_df
```

##### REMOVE ALL ROWS WITH NO DATA

```python
#Filter out those rows which does not contain any data 
AGrande_Tweets_df = AGrande_Tweets_df.dropna() 

#Confirm rows are dropped
shape = AGrande_Tweets_df.shape
print(f"The dataframes 'shape' is represented as: {shape}")
```

##### CORRECT THE INDEX FOR VISUAL PURPOSES

```python
#The original dataframe had an index, so re-arrange the index 
AGrande_Tweets_df.index = [x for x in range(1, len(AGrande_Tweets_df.values)+1)]
AGrande_Tweets_df.index.name = 'Index'
```

```python
#Print the dataframe
AGrande_Tweets_df
```
