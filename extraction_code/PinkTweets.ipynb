```python
#Import Dependencies
import pandas as pd
```

```python
#Create reference to CSV file
csv_path = "PinkTweets.csv"

#Import the CSV into a pandas DataFrame
Pink_Tweets_df = pd.read_csv(csv_path, low_memory = False)
Pink_Tweets_df
```

```python
#Show the 'shape' of the dataframe 
shape = Pink_Tweets_df.shape
print(f"The dataframes 'shape' is represented as: {shape}")
```

##### COLLECT ONLY REQUIRED COLUMNS

```python
#Select only columns we need
Pink_Tweets_df = Pink_Tweets_df[["Username", "User Id", "Tweet Type", "Tweet Content", "Client", "Retweets received",
                                     "Likes received", "Tweet Posted Time"]]

#Show the 'shape' of the dataframe
shape = Pink_Tweets_df.shape
print(f"The dataframes 'shape' is represented as: {shape}")
```

```python
#Print the dataframe
Pink_Tweets_df
```

##### REMOVE ALL ROWS WITH NO DATA

```python
#Filter out those rows which does not contain any data 
Pink_Tweets_df = Pink_Tweets_df.dropna() 

#Confirm rows are dropped
shape = Pink_Tweets_df.shape
print(f"The dataframes 'shape' is represented as: {shape}")
```

##### SEPERATE THE DATE AND TIME

```python
#Convert the 'Tweet Posted Time' column to string (from TimeStamp due to previous function to_datetime used)
Pink_Tweets_df['Tweet Posted Time'] = Pink_Tweets_df['Tweet Posted Time'].astype(str)

#Convert column to list
dt_list = Pink_Tweets_df["Tweet Posted Time"].tolist()

#Clean up the list to only show yyyy-mm-dd format
date = []
for x in dt_list:
    date.append(x.split(" ")[0])

#Add the formated lists into the dataframe as new columns
Pink_Tweets_df["Date"] = date
```

```python
#Convert the 'Tweet Posted Time' column to string (from TimeStamp due to previous function to_datetime used)
Pink_Tweets_df['Tweet Posted Time'] = Pink_Tweets_df['Tweet Posted Time'].astype(str)

#Convert column to list
dt_list = Pink_Tweets_df["Tweet Posted Time"].tolist()

#Clean up the list to show time
time = []
for x in dt_list:
    time.append(x.split(" ")[1])

#Add the formated lists into the dataframe as new columns
Pink_Tweets_df["Time"] = time
```

```python
#Drop the 'Tweet Posted Time' column as it is now replaced with a 'Date' and 'Time' column
Pink_Tweets_df = Pink_Tweets_df.drop(['Tweet Posted Time'], axis = 1)

#Show the 'shape' of the dataframe
shape = Pink_Tweets_df.shape
print(f"The dataframes 'shape' is represented as: {shape}")
```

```python
#Print the dataframe
Pink_Tweets_df
```

##### ADD A GENDER AND VOCATION COLUMN TO THE DATAFRAME AND FILL

```python
#Create an Gender column for Gender
Pink_Tweets_df['Gender'] = "Female"

#Create an Gender column for Gender
Pink_Tweets_df['Vocation'] = "Music"

#Show the 'shape' of the dataframe
shape = Pink_Tweets_df.shape
print(f"The dataframes 'shape' is represented as: {shape}")
```

##### CORRECT THE INDEX FOR VISUAL PURPOSES

```python
#The original dataframe had an index, so re-arrange the index 
Pink_Tweets_df.index = [x for x in range(1, len(Pink_Tweets_df.values)+1)]
Pink_Tweets_df.index.name = 'Index'
```

```python
#Print the dataframe
Pink_Tweets_df
```

##### SAVE DATAFRAME TO CSV

```python
#Save dataframe as csv file
Pink_Tweets_df.to_csv('Pink_Tweets_df.csv') 
```
