```python
# Set Up PyMongo
import pymongo
import pandas as pd
from flask import jsonify

client = pymongo.MongoClient("mongodb+srv://AtlasTwitter:1FineTwitterApp!@twittercluster.ycq9k.mongodb.net/")
mongo_db = client["Tweets_DB"]
mongo_collection = mongo_db["Combined_Tweets"]

# Define A Person to Query
# Searches for non-empty hashtag results 
QueryIdentity = "Jimmy Fallon"
testoutput =  mongo_collection.find( { "Identity": QueryIdentity, "Hashtags(#)": {"$ne" : '[]'} },{ "_id": 0, "Hashtags(#)": 1} ).limit(50)

test_df = pd.DataFrame(list(testoutput))
test_df.head(50)
```

```python
eval(test_df['Hashtags(#)'][0])
```

```python
import numpy as np 
# [eval(x) for x in test_df["Hashtags(#)"]]
testarray = [eval(x) for x in test_df["Hashtags(#)"]]
testarray
```

```python
flatten_list = []
for l1 in testarray:
    for l2 in l1:
        l2 = l2.replace('"','')
        flatten_list.append(l2)
        
flatten_list
```
