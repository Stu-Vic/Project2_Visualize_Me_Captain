```python
# Set Up PyMongo
import pymongo
import pandas as pd

client = pymongo.MongoClient("mongodb+srv://AtlasTwitter:1FineTwitterApp!@twittercluster.ycq9k.mongodb.net/")
mongo_db = client["Tweets_DB"]
mongo_collection = mongo_db["Combined_Tweets"]

# Define A Person to Query
# Searches for non-empty hashtag results 
QueryIdentity = "Jimmy Fallon"
testoutput =  mongo_collection.find( { "Identity": QueryIdentity, "Hashtags(#)": {"$ne" : '[]'} },{ "_id": 0, "Hashtags(#)": 1} ).limit(20)

test_df = pd.DataFrame(list(testoutput))
test_df.head(20)
```

```python
# Set up Pandas, Query Atlas for ALL results matching the twitter Identity 

identity_df = pd.DataFrame(list(mongo_collection.find({"Identity": QueryIdentity})))
```

```python

identity_df.head()
```

```python
identity_df['Time'] = pd.to_datetime(identity_df["Time"],format='%H:%M:%S')
# mega_df["Time"] = mega_df["Time"].dt.time
identity_df["Date"] = pd.to_datetime(identity_df["Date"],format="%Y-%M-%d")
```

```python
identity_df.head()
```

```python
identity_df.groupby(pd.Grouper(key='Date', freq='D'))
identity_df.mean()
```
